### 알고리즘 예제 6.6, 6.7, 6.8, 6.9
##### 문제 정의 : 선형 조사법의 삽입,연산,탐색을 사용한 알고리즘
##### 알고리즘 설명 : 해시 테이블에서 충돌을 해결하는 기법 중 하나이며 해시 함수를 사용하여 키를 해시값으로 변환한 다음, 해당 해시값에 해당하는 버킷에 키-값 쌍을 저장하는데, 이때 다른 키-값 쌍과 충돌이 발생할 수 있다.
##### 손으로 푼 예제 : 
##### 알고리즘 개요 : 삽입 연산,탑색 연산,삭제 연산을 사용하여 선형조사로 탐색한다.
