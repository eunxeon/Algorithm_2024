1) 문제 정의 (MD) : 허프만 트리 생성 알고리즘

2) 알고리즘 설명  (MD) : 허프만 코드 자체로는 빈도수를 통해 압축하여 저장되는 용량이 줄이거나 전체 데이터의 크기를 줄이고자 사용된다.

4) 코드 개요(입력 변수, 출력, 함수 설명) (MD) :
 - 입력 변수 : 노드 이름, 빈도수
 - 출력 : 가장 작은 빈도들부터 합쳐진 결과(단, 예시로 H가 3으로 가장 작다고 했을 때 합해진 결과가 다른 노드보다 크다면 합쳐진 노드의 순서가 뒤로 밀려날 수 있다)
 - 함수 설명 : make_heap_tree :> 힙을 사용하여 노드를 차례대로 힙에 넣고 가장 작은 순서대로 빼내어 더한 값을 다시 힙에 넣는다. 최종적으로 완성된 허프만 트리가 출력된다.

```python
# 5) 코드 (PY,C) : 
import heapq
def make_heap_tree(freq, label):
    n = len(freq)
    h = []
    for i in range(n):
        heapq.heappush(h, (freq[i], label[i]))
    
    for i in range(1, n):
        e1 = heapq.heappop(h)
        e2 = heapq.heappop(h)
        heapq.heappush(h, (e1[0]+e2[0], e1[1]+e2[1]))
        print(e1, "+", e2)
    
    print(heapq.heappop(h))
```

```python
# 6) 테스트 코드  (PY,C) , 입력을 바꾸어서 : 
label = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
freq = [  24,   3,   8,  10,  33,   6,   4,  12]

make_heap_tree(freq, label)
```
