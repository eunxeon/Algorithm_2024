### 알고리즘 예제 5.3
##### 문제 정의 : 정렬
##### 알고리즘 설명 : 리스트에서 피벗을 선택하고 이를 기준으로 리스트를 2개의 부분 리스트로 나누며, 각 부분 리스트를 더 이상 나눌 수 없을 때 까지 반복
##### 손으로 푼 예제 : 
##### 알고리즘 개요 : (입력) data / (출력) Oreginal,QuickSort / (함수) partition(A, left, right),quick_sort(A, left, right)
##### 수행 결과 : 
##### 복잡도 분석 : 최악의 경우O(n^2) 평균적인 경우O(nlog2 n) 최선의 경우O(n log2 n)
###### 보통 피벗의 위치를 반환하는 partition() 연산이 0~n-1을 동일한 확률로 반환한다는 가정이 많이 이용되는데, 만약 피벗의 위치를 s라고 한다면 평균적인 복잡도는 O(nlog2 n)이다.

```python
def quick_sort(A, left, right):         #A[left..right]를 오름차순으로 정렬
    if left<right:                      #정렬 범위가 2개 이상인 경우
        mid = partition(A, left, right) #좌우로 분할
        quick_sort(A, left, mid-1)      #왼쪽 부분리스트를 퀵 정렬
        quick_sort(A, mid+1, right)     #오른쪽 부분리스트를 퀵 정렬
```

```python
def partition(A, left, right):
    low = left + 1
    high = right
    pivot = A[left]
    while (low <= high) :
        while low <= right and A[low] <= pivot : low += 1
        while high >= left and A[high] > pivot : high -= 1

        if low < high:
            A[low], A[high] = A[high], A[low]
    
    A[left], A[high] = A[high], A[left]
    return high
def quick_sort(A, left, right):         
    if left<right:                      
        mid = partition(A, left, right)
        quick_sort(A, left, mid-1)      
        quick_sort(A, mid+1, right)    
data= [5, 3, 8, 4, 9, 1, 6, 2, 7]
print("Original :", data)
quick_sort(data, 0, len(data)-1)
print("QuickSort :", data)
def partition(A, left, right):
    low = left + 1
    high = right
    pivot = A[left]
    while (low <= high) :
        while low <= right and A[low] <= pivot : low += 1
        while high >= left and A[high] > pivot : high -= 1

        if low < high:
            A[low], A[high] = A[high], A[low]
    
    A[left], A[high] = A[high], A[left]
    return high
def quick_sort(A, left, right):         
    if left<right:                      
        mid = partition(A, left, right)
        quick_sort(A, left, mid-1)      
        quick_sort(A, mid+1, right)    
data= [5, 3, 8, 4, 9, 1, 6, 2, 7]
print("Original :", data)
quick_sort(data, 0, len(data)-1)
print("QuickSort :", data)
```
