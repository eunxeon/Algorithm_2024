######  2) 알고리즘 설명  (MD) : 정렬하기 쉽도록 리스트를 쪼갠 후 정렬 이후에 다시 합쳐서 정렬된 리스트를 획득하는 알고리즘이다. 
######  3) 손으로 푼 예제 (MD, 이미지 삽입) : 
######  4) 코드 개요(입력 변수, 출력, 함수 설명) (MD) : 
######  입력 변수 : 리스트 A, 시작 인덱스, 끝 인덱스
######  출력 변수 : 정렬된 리스트 A
######  함수 설명 : merge_sort 함수 >> 데이터 리스트들을 균등하게 분할하는 함수
######              merge 함수 >> 정렬된 두 리스트를 병합하는 함수

```python
# 5) 코드 (PY,C) :
# 알고리즘 5.1 병합 정렬
def merge_sort(A, left, right):
    if left < right:
        mid = (left + right) // 2 #리스트의 균등 분할
        merge_sort(A, left, mid)
        merge_sort(A, mid+1, right)
        merge(A, left, mid, right)
```

```python
# 5) 코드 (PY,C) :
# 알고리즘 5.2 병합 알고리즘(정렬된 두 리스트의 병합)
def merge(A, left, mid, right):
    k = left
    i = left
    j = mid + 1
    while i<=mid and j <=right:
        if A[i] <= A[j]:
            sorted[k] = A[i]
            i, k = i+1, k+1
        else :
            sorted[k] = A[j]
            j, k = j+1, k+1
    
    if i > mid:
        sorted[k:k+right-j+1] = A[j:right+1]
    else :
        sorted[k:k+mid-i+1] = A[i:mid+1]

    A[left:right+1] = sorted[left:right+1]
```

```python
######  6) 테스트 코드  (PY,C) , 입력을 바꾸어서 : 
data = [ 5, 3, 8, 4, 9, 1, 6, 2, 7 ]
sorted = [0] * len(data)
print("Original : ", data)
merge_sort(data, 0, len(data)-1)
print("MergeSort : ", data)
```
